##  AF_wo_INFO_001 after applying --mind 0.05 to remove 279 high missingness people


```r
af_mind_lmiss=fread("mind_filter.missing.lmiss")
```

```r
af_mind_imiss=fread("mind_filter.missing.imiss")
```

```r
head(af_mind_lmiss)
```

```r
head(af_mind_imiss)
```

```r
ggplot(data=af_mind_imiss,
      aes(x=af_mind_imiss$F_MISS))+
geom_histogram(bins=100)+
geom_vline(xintercept = 0.05,color='red')+
ggtitle("AF --mind 0.05 imiss")
```

```r
ggplot(data=af_mind_lmiss,
      aes(x=af_mind_lmiss$F_MISS))+
geom_histogram(bins=100)+
geom_vline(xintercept = 0.05,color='red')+
ggtitle("AF --mind 0.05 lmiss")
```

```r
ggplot(data=af_mind_lmiss,
      aes(x=af_mind_lmiss$F_MISS))+
geom_histogram(bins=100)+
geom_vline(xintercept = 0.05,color='red')+
geom_vline(xintercept = 0.01,color='blue')+
ggtitle("AF --mind 0.05 lmiss")+
xlim(c(0,0.501))
```
